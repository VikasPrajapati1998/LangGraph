## **<font color="red">Long Term Memory</font>**

````python
from langgraph.store.memory import InMemoryStore
````

````python
store = InMemoryStore()
````

### **Creating Memories**
- (user, u1), (user, u2) ... 
- (user, u1, profile), (user, u2, profile)
- (user, u1, preference), (user, u2, preference)
- Memory
```
user
|- u1 -> User 1
|   |- profile
|   |- preference
|
|- u2 -> User 2
...
```

````python
namespace1 = ("user", "u1")

# Adding Memories
store.put(namespace1, "1", {"data": "User likes pizza"})
store.put(namespace1, "2", {"data", "User prefers dark mode"})
````

````python
namespace2 = ("user", "u2")

# Adding Memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data", "User prefers grid style navigation"})
````

### **Retrieving Memories**

````python
# namespace 1
u1_data1 = store.get(namespace1, "1")
print(u1_data1)
u1_data2 = store.get(namespace1, "2")
print(u1_data2)
````

````python
# namespace 2
u2_data1 = store.get(namespace2, "1")
print(u2_data1)
u2_data2 = store.get(namespace2, "2")
print(u2_data2)
````

### **Retrieving all Memories**

````python
# namespace 1
u1_items = store.search(namespace1)
for item in u1_items:
    print(item.value)
````

````python
# namespace 2
u2_items = store.search(namespace2)
for item in u2_items:
    print(item.value)
````
