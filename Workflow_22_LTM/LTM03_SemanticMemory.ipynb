```python
from langgraph.store.memory import InMemoryStore
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(model="nomic-embed-text:v1.5")
store = InMemoryStore(index={"embed": embedding_model, "dims": 1536})
```

```python
namespace1 = ('users', 'u1')

store.put(namespace1, "1", {"data": "User prefers concise answers over long explanation"})
store.put(namespace1, "2", {"data": "User likes examples in Python"})
store.put(namespace1, "3", {"data": "User usually works late at night"})
store.put(namespace1, "4", {"data": "User prefers dark mode in applications"})
store.put(namespace1, "5", {"data": "User is learning machine learning"})
store.put(namespace1, "6", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace1, "7", {"data": "User prefers step-by-step reasoning"})
store.put(namespace1, "8", {"data": "User is based in India"})
store.put(namespace1, "9", {"data": "User likes real-world analogies"})
store.put(namespace1, "10", {"data": "User prefers bullet points over paragraphs"})
```

```python
items = store.search(namespace1, query="What is the user currently learning", limit=2)

for item in items:
    print(item.value)
```
