# **<font color="red">Multiprocessing in Python</font>**
## **<font color="blue">Introduction</font>**
- Parallel processing underpins the acceleration of execution times of codes that previously required a significant amount of time to execute.
- Parallel processing is a technique that involves running multiple tasks simultaneously to increase efficiency.
- This can be achieved either by running tasks on different cores of CPUs or utilizing CPU cycles withing the same core.
- There are two primary types of parallel processing:
  1. **Multiprocessing**
  2. **Multithreading**
- **<font color="orange">Multiprocessing:</font>**
  - Imagine you have a CPU with 4 cores. Think of each core as an individual worker and the operating system as the supervisor i.e., Each core executes tasks assigned by the operating system.
  - In this scenario, a "process" is a sequence of commands provided to a core. Each core follows these instructions to perform a task.
  - Multiprocessing involves distributing tasks among different cores, allowing them to work on multiple tasks simultaneously. This is akin to having multiple workers perform different tasks at the same time.
- **<font color="orange">Multithreading:</font>**
  - In multithreaded approach, consider a thread as a conveyor belt. The belt carries a series of tasks (code sequences) to the worker (core) one by one.
  - Threads are like instructions that guide a core on what to do. While one thread is being executed by a core, another can be prepared to execute on the same core afterward.
  - This allows multiple tasks to be processed sequentially on a single core. It is like having a single worker (core) work on different tasks in a continous stream.

## **<font color="blue">Multiprocessing in Python</font>**
- Python is a versatile programming language; however, its default single-threaded execution can lead to slow performance for long-running tasks.
- To overcome this limitation and speed up processing, Python offers built-in multiprocessing capabilities, which allow codes to run on multiple processes and leverage the power of multiple CPU cores.

# **<font color="red">`multiprocessing` --- Process-based parallelism</font>**
## **<font color="blue">Introduction</font>**
- Multiprocessing is a package that support spawning processes using an API similar to the threading module. The multiprocessing package offers both local and remote concurrency, effectively side-stepping the GIL(Global Interpreter Lock) by using subprocesses instead of threads. Due to this, the multiprocessing module allows the programmer to fully leverage multiple processors on a given machine. It runs on both POSIX and Windows.
- The ***threading*** module provides a way to run multiple threads concurrently within a single process. It allows for the creation and management of threads, making it possible to execute tasks in parallel, sharing memory space. Threads are particulary useful when tasks are I/O bound, such as file operations or making network requests, where much of the time is spent waiting for external resources.
```python
import threading
import time

def task(name):
    for i in range(5):
        print(f"{name} is running {i}")
        time.sleep(1)

# Create threads
t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))

# Start threads
t1.start()
t2.start()

# Wait for threads to finish
t1.join()
t2.join()

print("All threads finished")
```
- The multiprocessing module also introduces the **Pool** object which offers a convenient means of parallelizing the execution of a function across multiple input values, distributing the input data across processes (data parallelist).
- A process ***pool*** object which controls a pool of worker processes to which jobs can be submitted. It supports asynchronous results with timeouts and callbacks and has a parallel map implementation.
  - *processes* is the number of worker processes to use. If *processes* is `None` then the number returned by `os.process_cpu_count()` is used.
  - If *initializer is not `None` then each worker process will call `initializer(*initargs)` when it starts.
  - *maxtasksperchild* is the number of tasks a worker process can complete before it will exit and be replaced with a fresh worker process, to enable unused resources to be freed. The default *maxtasksperchild* is `None`, which means worker processes will live as long as the pool.
```python
import multiprocessing
import time

def task(name):
    for i in range(5):
        print(f"{name} running {i}")
        time.sleep(1)

if __name__ == "__main__":
    # Create processes
    p1 = multiprocessing.Process(target=task, args=("Process-1",))
    p2 = multiprocessing.Process(target=task, args=("Process-2",))

    # Start processes
    p1.start()
    p2.start()

    # Wait for them to finish
    p1.join()
    p2.join()

    print("All processes finished")
```

```python
import multiprocessing

def square(n):
    return n * n

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5]

    # Create a pool of worker processes
    with multiprocessing.Pool(3) as pool:
        results = pool.map(square, numbers)

    print("Results:", results)
```

````python
# multi-threading
import threading
import time

def task(name):
    for i in range(5):
        print(f"{name} is running {i}")
        time.sleep(1)

# Create threads
t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))

# Start threads
t1.start()
t2.start()

# Wait for threads to finish
t1.join()
t2.join()

print("All threads finished")
````

````python
# multi-processing
import multiprocessing
import time

def task(name):
    for i in range(5):
        print(f"{name} running {i}")
        time.sleep(1)

if __name__ == "__main__":
    # Create processes
    p1 = multiprocessing.Process(target=task, args=("Process-1",))
    p2 = multiprocessing.Process(target=task, args=("Process-2",))

    # Start processes
    p1.start()
    p2.start()

    # Wait for them to finish
    p1.join()
    p2.join()

    print("All processes finished")
````

````python
# multi-processing with Pool
import multiprocessing

def square(n):
    return n * n

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5]

    # Create a pool of worker processes
    with multiprocessing.Pool(3) as pool:
        results = pool.map(square, numbers)

    print("Results:", results)
````
